## 微分方程

微分方程，其实就是普通方程。普通方程，也就是方程，或者方程组，描述了一些量之间的不变性。

也就是：

$$
f(x, y, z) = g(x, y, z)\quad,\text{or}\quad F(x, y, z) = 0
$$

或者这些量是向量，那么就是向量方程：

$$
\vec{f}(\vec{r}) = \vec{g}(\vec{r})\quad,\text{or}\quad \vec{F}(\vec{r}) = \vec{0}
$$

其中，$\vec{r} = [x, y, z]^\text{T}$ 是空间中的点，$\vec{f}$ 和 $\vec{g}$ 是向量场。

微分方程就是平凡的方程，只不过，方程中出现了变量的变化率和变化率的变化率。

考虑变化率构成的方程，就是一阶微分方程；考虑变化率的变化率构成的方程，就是二阶微分方程。

再强调一遍，微分方程就是方程。

那么微分方程涉及的变化率、变化率的变化率，都是什么？

## 向量微分学-梯度

对于标量函数 $f(x, y, z)$，变化率就是梯度向量：

$$
\text{grad} f = \nabla f = \left[ \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \right]^\text{T}
$$

对于向量函数 $\vec{f}(x, y, z) = [f_1(x, y, z), f_2(x, y, z), f_3(x, y, z)]^\text{T}$，变化率也就是梯度矩阵：

$$
\text{grad} \vec{f} = \nabla \vec{f} = \left[ 
    \begin{array}{ccc}
        \frac{\partial f_1}{\partial x} & \frac{\partial f_1}{\partial y} & \frac{\partial f_1}{\partial z} \\
        \frac{\partial f_2}{\partial x} & \frac{\partial f_2}{\partial y} & \frac{\partial f_2}{\partial z} \\
        \frac{\partial f_3}{\partial x} & \frac{\partial f_3}{\partial y} & \frac{\partial f_3}{\partial z}
    \end{array}
    \right]
$$


梯度的方向是函数变化率最大的方向，梯度的大小是函数变化率的最大值。

梯度有非常明确的物理意义。

梯度将一个标量场转化为一个向量场，将一个向量场转化为一个2阶张量场。

$$
\text{grad} : (\mathbb{R}^3 \mapsto \mathbb{R}) \mapsto (\mathbb{R}^3 \mapsto \mathbb{R}^3)
$$


$$
\text{grad} : (\mathbb{R}^3 \mapsto \mathbb{R}^3) \mapsto \left(\mathbb{R}^3 \mapsto (\mathbb{R}^3 \otimes \mathbb{R}^3)\right)
$$


对于三维空间（也就是我们身处其中的空间）中的场量，也就是一个 $\mathbb{R}^3 \mapsto \mathbb{R}$ 的函数 $f(x,y,z)$，梯度将这个函数转化为一个向量场。这个向量场也就是一个定义在 $\mathbb{R}^3$ 上的 $\nabla f = \vec{p}(x,y,z)\in \mathbb{R}^3$ 的函数，在每个空间点上，这个向量的方向就是函数变化率最大的方向，这个向量的模就是函数变化率的最大值。







## 向量微分学-散度

散度，跟梯度不同。散度将一个向量转化为一个标量。

$$
\text{div} : (\mathbb{R}^3 \mapsto \mathbb{R}^3) \mapsto (\mathbb{R}^3 \mapsto \mathbb{R})
$$

$$
\text{div} \vec{f} = \nabla \cdot \vec{f} = \frac{\partial f_1}{\partial x} + \frac{\partial f_2}{\partial y} + \frac{\partial f_3}{\partial z}
$$

这里：

$$
\vec{f} = \left[
    \begin{array}{c}
        f_1 \\
        f_2 \\
        f_3
    \end{array}
\right]
$$

## 向量微分学-梯度的散度

当梯度和散度结合在一起，就得到了拉普拉斯算子。这个算子将一个标量场转化为一个标量场。

$$
\text{laplacian} : (\mathbb{R}^3 \mapsto \mathbb{R}) \mapsto (\mathbb{R}^3 \mapsto \mathbb{R})
$$

它的定义就是梯度的散度。

$$
\text{laplacian} f \equiv \nabla \cdot (\nabla f) = \nabla^2 f
$$
